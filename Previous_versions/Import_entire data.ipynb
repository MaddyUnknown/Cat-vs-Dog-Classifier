```python
import os
import random
from PIL import Image
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

## Prepare training data

```python
x = np.array([])
y = np.array([])
path = r'dataset\training_set'
classes = ['cats', 'dogs']
for i in range(len(classes)):
    path_in = os.path.join(path, classes[i])
    for j in os.listdir(path_in):
        img = Image.open(os.path.join(path_in, j))
        img = img.resize((120,120), Image.ANTIALIAS)
        img = np.asarray(img)
        img = np.rollaxis(img, 2, 0)
        img = img.reshape((1,3,120,120))
        if(len(x)==0):
            x = img
            y = np.array([[i]])
        else:
            x  = np.append(x, img, axis=0)
            y  = np.append(y, [[i]], axis=0)
        print('Image: ', j, '            Size: ', img.shape)

permutation = [i for i in range(len(x))]
random.shuffle(permutation)

x = x[permutation]
y = y[permutation]

print("\n\n------------------------------------------- Data imported ----------------------------------------------")
print("Number of data: ", len(x))
print("Shapes: x =", x.shape, " y =", y.shape)

import winsound
winsound.Beep(325, 900)
```

## Model

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 5)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(2, stride=2)
        self.conv2 = nn.Conv2d(16, 8, 3)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(2, stride=2)
        self.conv3 = nn.Conv2d(8, 10, 3)
        self.relu3 = nn.ReLU()
        self.pool3 = nn.MaxPool2d(2, stride=2)
        self.fc1 = nn.Linear(1690, 120)
        self.relu_fc1 = nn.ReLU()
        self.fc2 = nn.Linear(120, 80)
        self.relu_fc2 = nn.ReLU()
        self.fc3 = nn.Linear(80,1)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        out = self.pool1(self.relu1(self.conv1(x)))
        out = self.pool2(self.relu2(self.conv2(out)))
        out = self.pool3(self.relu3(self.conv3(out)))
        out = out.view(-1, self.num_flat_features(out))
        out = self.relu_fc1(self.fc1(out))
        out = self.relu_fc2(self.fc2(out))
        out = self.sigmoid(self.fc3(out))
        
        return out
    
    def test(self, x):
        out = self.conv1(x)
        return out
        
        
    def num_flat_features(self, x):
        size = x.size()[1:]  # all dimensions except the batch dimension
        num_features = 1
        for s in size:
            num_features *= s
        return num_features
        
```

```python
model = Net().cuda()
print(model)
```

```python
import time

itter = 10
criteria = nn.BCELoss()
costs = []
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)

x_train = torch.Tensor(x[:20]).float().cuda()
y_train = torch.Tensor(y[:20]).float().cuda()
x_train.requires_grad_(True)
t = time.time()
for i in range(itter):
    output = model(x_train)
    optimizer.zero_grad()
    loss = criteria(output, y_train)
    costs.append(loss)
    loss.backward()
    optimizer.step()
    if(i%10==0):
        print('epoch {}, loss {}'.format(i, loss.data))
print("Time taken: ", time.time()-t, "\n")
```
